```python
import os
import cv2
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import joblib
from tensorflow.keras.applications import VGG16

# Function to extract features from VGG16 model
def extract_features(model, image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (224, 224))
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize pixel values
    features = model.predict(img)
    return features.flatten()

# Load a pre-trained model (VGG16 in this example)
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the layers of the pre-trained model
for layer in base_model.layers:
    layer.trainable = False

# Paths to the train dataset
train_dataset_path = "C:\\Users\\appin\\Desktop\\Leaf-Classification-Medicinal-master\\Train"

# Initialize lists to store features and labels
features = []
labels = []

# Load image file paths and labels from the train dataset
for class_name in class_names:
    class_path = os.path.join(train_dataset_path, class_name)
    img_files = os.listdir(class_path)
    
    for file in img_files:
        img_path = os.path.join(class_path, file)
        features.append(extract_features(base_model, img_path))
        labels.append(class_name)

# Convert labels to numerical values
label_encoder = LabelEncoder()
labels_encoded = label_encoder.fit_transform(labels)

# Train an SVM classifier
svm_classifier = SVC(kernel='linear', C=1.0)
svm_classifier.fit(features, labels_encoded)

# Save the SVM classifier and label encoder to files
joblib.dump(svm_classifier, 'svm_classifier.pkl')
joblib.dump(label_encoder, 'label_encoder.pkl')
```

```python
import cv2
import numpy as np
import joblib
import os
from tensorflow.keras.applications import VGG16
from sklearn.metrics import accuracy_score

class_names = ['Arive-Dantu', 'Basale', 'Betel', 'Crape_Jasmine', 'Curry', 'Drumstick', 'Fenugreek', 'Guava', 'Hibiscus', 'Indian_Beech', 'Indian_Mustard', 'Jackfruit', 'Jamaica_Cherry-Gasagase', 'Jamun', 'Jasmine', 'Karanda', 'Lemon', 'Mango', 'Mexican_Mint', 'Mint', 'Neem', 'Oleander', 'Parijata', 'Peepal', 'Pomegranate', 'Rasna', 'Rose_Apple', 'Roxburgh_fig', 'Sandalwood', 'Tulsi']

# Function to extract features from VGG16 model
def extract_features(model, image_path):
    img = cv2.imread(image_path)
    
    # Check if the image is None
    if img is None:
        print(f"Failed to read image: {image_path}")
        return None
    
    print(f"Image Path: {image_path}")
    print(f"Image Shape: {img.shape}")
    
    img = cv2.resize(img, (224, 224))
    
    print(f"Resized Image Shape: {img.shape}")
    
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize pixel values
    features = model.predict(img)
    return features.flatten()


# Load the VGG16 model
base_model = VGG16(weights='imagenet', include_top=False)

# Load the trained SVM classifier and label encoder
svm_classifier = joblib.load('svm_classifier.pkl')
label_encoder = joblib.load('label_encoder.pkl')

# Paths to the test dataset
test_dataset_path = "C:\\Users\\deera\\OneDrive\\Desktop\\capstone_proj\\Medicinal-Plant-Leaf-DetectionOLD\\Test"

# Initialize lists to store test features and true labels
test_features = []
true_labels = []

# Load image file paths and true labels from the test dataset
for class_name in class_names:
    class_path = os.path.join(test_dataset_path, class_name)
    img_files = os.listdir(class_path)
    
    for file in img_files:
        img_path = os.path.join(class_path, file)
        features = extract_features(base_model, img_path)
        
        # Skip None elements (failed to read image)
        if features is not None:
            test_features.append(features)
            true_labels.append(class_name)

# Convert true labels to numerical values
true_labels_encoded = label_encoder.transform(true_labels)

# Make predictions on the test set
predicted_labels_encoded = svm_classifier.predict(test_features)

# Convert predicted labels back to class names
predicted_labels = label_encoder.inverse_transform(predicted_labels_encoded)

# Evaluate the accuracy
accuracy = accuracy_score(true_labels, predicted_labels)
print(f'Test Accuracy: {accuracy * 100:.2f}%')
```

```python
import cv2
import numpy as np
import joblib
from tensorflow.keras.applications import VGG16

# Function to extract features from VGG16 model
def extract_features(model, image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (224, 224))
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize pixel values
    features = model.predict(img)
    return features.flatten()

# Function to predict the class of a single leaf image
def predict_leaf_class(image_path, svm_classifier, label_encoder, base_model):
    features = extract_features(base_model, image_path)
    features = np.array(features).reshape(1, -1)
    class_index = svm_classifier.predict(features)[0]
    predicted_class = label_encoder.classes_[class_index]
    return predicted_class

# Load the trained SVM classifier and label encoder
svm_classifier = joblib.load('svm_classifier.pkl')
label_encoder = joblib.load('label_encoder.pkl')

# Load a pre-trained model (VGG16 in this example)
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the layers of the pre-trained model
for layer in base_model.layers:
    layer.trainable = False

# Example usage:
test_image_path = r"C:\Users\deera\OneDrive\Desktop\capstone_proj\Medicinal-Leaf-Detection (3)\Medicinal-Leaf-Detection\upload\blurbasale.jpg"
predicted_class = predict_leaf_class(test_image_path, svm_classifier, label_encoder, base_model)
print(f'Predicted Class: {predicted_class}')
```
